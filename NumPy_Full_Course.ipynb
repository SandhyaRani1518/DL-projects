## 1) What is NumPy and Why use it

NumPy (Numerical Python) is a core Python library used for scientific computing.
- Fast and memory efficient compared to Python lists
- Provides ndarray: n-dimensional array object
- Supports vectorized operations (no Python loops)


## 2) Speed comparison: Python list vs NumPy array

```python
import numpy as np      #alias ---> np
import time

L = list(range(10000000))
start = time.time()
res = [x*2 for x in L]
list_time = time.time()-start

A = np.arange(10000000)
start = time.time()
res = A * 2
np_time = time.time()-start


list_time, np_time
```

## 4) Array attributes & Indexing

- `.shape`, `.size`, `.ndim`
- Indexing and slicing examples

```python
import numpy as np

arr = np.array([[10,20,30],[40,50,60],[70,80,90]])

arr_shape = arr.shape
arr_size = arr.size
arr_ndim = arr.ndim

# indexing and slicing example

item = arr[0,1]
sub = arr[1:,1:]
col2 = arr[:,1]
row3 = arr[2,:]

arr_shape, arr_size, arr_ndim, item, sub, col2, row3
```

## 5) Mathematical Operations (vectorized)

Add, multiply, power, sqrt, sum etc.

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

add = a + b
sub = a - b
mul = a * b
sq = a ** 2
sqrt_b = np.sqrt(b)
sum_a = np.sum(a)
```

## 8) Combining, Splitting & Sorting

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

concat = np.concatenate((a,b))
vstack = np.vstack((a,b))
hstack = np.hstack((a,b))

arr = np.array([10,210,30, 140,50,60])
splits = np.split(arr,2)
sorted_arr = np.sort(arr)

sorted_arr.tolist()
```

## 9) Advanced Tricks: Boolean indexing, unique, dot (matrix multiply)

```python
import numpy as np

arr = np.arange(10)

filtered = arr[arr>5]

arr1 = np.array([1,1,2,3,4, 4, 5, 5])

unique_vals = np.unique(arr1)

filtered, unique_vals

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

matmaul = np.dot(A,B)

matmaul.tolist()
```
